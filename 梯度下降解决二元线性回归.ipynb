```python
import numpy as np
```

```python
np.random.seed(1)
```

```python
x1 = np.random.normal(size = (1000, 1), scale = 1)
x2 = np.random.normal(size = (1000, 1), scale = 1)
```

# 定义函数

```python
y = 2 * x1 - 3 * x2 + 8
```

```python
All_data = np.concatenate((x1, x2, y), axis = 1)
Train_data = All_data[:700]
Test_data = All_data[700:]
```

## 梯度下降求解二元线性回归
## BGD


```python
theta1 = np.random.rand()
theta2 = np.random.rand()
lr = 0.001
```

```python
#偏置
theta0 = np.random.rand()
```

```python
#定义最大迭代次数
n = 1e4
```

```python
num = 1
theta0_list = []
theta1_list = []
theta2_list = []
while True:
    np.random.shuffle(Train_data)
    x_data = Train_data[:, :-1]
    y_data = Train_data[:, -1]
    m = len(x_data)
    for j in range(0,m):
        theta0 = theta0 - ((theta1*x_data[j,0]+theta2*x_data[j,1] + theta0)-y_data[j]) * lr *  1 / m
        theta1 = theta1 - x_data[j,0]*((theta1*x_data[j,0]+theta2*x_data[j,1]+theta0)-y_data[j]) * lr * 1 / m
        theta2 = theta2 - x_data[j,1]*((theta1*x_data[j,0]+theta2*x_data[j,1]+theta0)-y_data[j]) * lr * 1 / m
    loss = np.sum(0.5 * ((theta1 * x_data[:, 0] + theta2 * x_data[:, 1] + theta0 - y_data) ** 2))
    print("num = %d, theta0 = %f, theta1 = %f, theta2 = %f, loss = %f"%(num, theta0, theta1, theta2, loss))
    theta0_list.append(theta0)
    theta1_list.append(theta1)
    theta2_list.append(theta2)
    num += 1
    if num > n:
        break
            
```

```python
print("w1 = %f, w2 = %f, b = %f, loss = %f"%(theta1, theta2, theta0, loss))
```
